```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aqi=pd.read_csv("aqi.csv")
aqi.info()
```

```python
pop=pd.read_csv("population_projection.csv")
pop.info()
```

```python
aqi['date'] = pd.to_datetime(aqi['date'], format="%d-%m-%Y", errors='coerce')
```

```python
aqi['date'] = pd.to_datetime(aqi['date'], errors='coerce')
aqi_2024 = aqi[aqi['date'].dt.year == 2024]
```

```python
avg_aqi_per_state = aqi_2024.groupby('state')['aqi_value'].mean().reset_index()
avg_aqi_per_state.rename(columns={'aqi_value': 'avg_aqi_2024'}, inplace=True)
```

```python
pop_2024 = pop[(pop['year'] == 2024) & (pop['gender'].str.lower() == 'total')]
pop_2024 = pop_2024.groupby('state')['value'].sum().reset_index()
pop_2024.rename(columns={'value': 'population_2024'}, inplace=True)
pop_2024
```

```python
merged_df = pd.merge(avg_aqi_per_state, pop_2024, on='state')
merged_df
```

```python
# Step 7: Plot with dual axes
fig, ax1 = plt.subplots(figsize=(14, 7))

# Bar plot for population
ax1.bar(merged_df['state'], merged_df['population_2024'], color='skyblue', label='Population (2024)')
ax1.set_xlabel('State')
ax1.set_ylabel('Population (2024)', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')
ax1.set_xticklabels(merged_df['state'], rotation=45, ha='right')

# Line plot for AQI
ax2 = ax1.twinx()
ax2.plot(merged_df['state'], merged_df['avg_aqi_2024'], color='crimson', marker='o', label='Avg AQI (2024)')
ax2.set_ylabel('Average AQI (2024)', color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson')

# Title and legend
plt.title('State-wise Population and Average AQI in 2024')
fig.tight_layout()
plt.show()
```

```python
correlation = merged_df['population_2024'].corr(merged_df['avg_aqi_2024'])
print(f"Correlation between Population and AQI: {correlation:.2f}")
```

```python
merged_df.to_csv('Population_and_Average_AQI.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
poped=pd.read_csv("Indian_States_Population_Density.csv")
poped.head()
```

```python
popedsort = poped.sort_values("Population_Density_per_sq_km", ascending=False)
popedsort
```

```python
toppop=popedsort.head(10)
toppop
```

```python
top_state=toppop['State'].tolist()
top_state.append("Maharashtra")
print(top_state)
```

```python
aqi=pd.read_csv("aqi.csv")
aqi
```

```python
aqi1=aqi[aqi['state'].isin(top_state)]
print(aqi1.shape)
```

```python
aqi1['state'].unique()
```

```python
# Step 1: Clean any invisible whitespace (just to be safe)
aqi1['date'] = aqi1['date'].astype(str).str.strip()

# Step 2: Convert using the exact format
aqi1['date'] = pd.to_datetime(aqi1['date'], format='%d-%m-%Y', errors='coerce')

# Step 4: Verify a few dates
print(aqi1['date'].head())
```

```python
aqi1.loc[:, 'month'] = aqi1['date'].dt.strftime('%B')   # 'January', 'February', etc.
aqi1.loc[:, 'month_num'] = aqi1['date'].dt.month        # 1–12 for sorting
```

```python
aqi1
```

```python
monthly_aqi = (
    aqi1.groupby(['state', 'month', 'month_num'])['aqi_value']
    .mean()
    .reset_index()
)
monthly_aqi
```

```python
worst_months = (
    monthly_aqi.sort_values(['state', 'aqi_value'], ascending=[True, False])
    .groupby('state')
    .first()
    .reset_index()
)
worst_months
```

```python
month_counts = (
    worst_months['month']
    .value_counts()
    .sort_values(ascending=False)
    .reset_index()
    .rename(columns={'index': 'Month', 'month': 'Frequency'})
)
month_counts
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate N equally spaced colors from the viridis colormap
colors = [plt.cm.viridis(i / len(month_counts)) for i in range(len(month_counts))]

# Create the pie chart
plt.figure(figsize=(4, 4))
plt.pie(
    month_counts['count'], 
    labels=month_counts['Frequency'], 
    autopct='%1.1f%%', 
    startangle=140,
    colors=colors
)

plt.title("Most Frequent Worst AQI Months across Top 10 States (by Area Coverage)")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
month_counts.to_csv('filename.csv', index=False)
```
